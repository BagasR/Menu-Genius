```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import joblib
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("/content/data_fix.csv")
```

```python
df
```

```python
df.isna().sum()
```

```python
def cari_Null(data):
    kolom_dgn_null = data.isna().sum()
    kolom_dgn_null = kolom_dgn_null[kolom_dgn_null > 0].sort_values(ascending=False)
    return kolom_dgn_null
```

```python
cari_Null(df)
```

```python
df.describe()
```

```python
def rentang_harga(Harga):
  if Harga >= df['Harga_Rata-rata'].min() and Harga <= df['Harga_Rata-rata'].quantile(q=0.25):
    return 1
  elif Harga > df['Harga_Rata-rata'].quantile(q=0.25) and Harga <= df['Harga_Rata-rata'].median():
    return 2
  elif Harga > df['Harga_Rata-rata'].median() and Harga <= df['Harga_Rata-rata'].quantile(q=0.75):
    return 3
  elif Harga > df['Harga_Rata-rata'].quantile(0.75) and Harga <= df['Harga_Rata-rata'].max():
    return 4
```

```python
df['Rentang_Harga_Rata-Rata'] = df['Harga_Rata-rata'].apply(rentang_harga)
```

```python
df['Rentang_Harga_Rata-Rata'].value_counts()
```

```python
def Kec_Angka(Kecamatan):
  if Kecamatan == 'Syiah Kuala':
    return 1
  elif Kecamatan == 'Kuta Alam':
    return 2
  elif Kecamatan == 'Banda Raya':
    return 3
  elif Kecamatan == 'Meuraxa':
    return 4
```

```python
df['kecamatan'] = df.Kecamatan.apply(Kec_Angka)
```

```python
df.kecamatan.value_counts()
```

```python
def Kab_Angka(Kabupaten):
  if Kabupaten == 'Banda Aceh':
    return 1
  else:
    return 2
```

```python
df['kabupaten'] = df.Kabupaten.apply(Kab_Angka)
```

```python
df.kabupaten.value_counts()
```

```python
def CafeAtauRestoran(CF):
  if CF == 'restoran':
    return 1
  else:
    return 2
```

```python
df['CR_Angka'] = df['Cafe/restoran'].apply(CafeAtauRestoran)
```

```python
df.CR_Angka.value_counts()
```

```python
df
```

```python
df.describe()
```

```python
X =  df[['Harga_Rata-rata','rating','kecamatan','kabupaten','CR_Angka']]
y =  df['Rentang_Harga_Rata-Rata']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
imputer = SimpleImputer(strategy='mean')  # Gantilah 'mean' sesuai kebutuhan Anda
X_train = imputer.fit_transform(X_train)
```

```python
X_test = X_test.dropna()
y_test = y_test.loc[X_test.index]
```

```python
model = DecisionTreeClassifier()
model = model.fit(X_train, y_train)
```

```python
accuracy_train = model.score(X_train, y_train)
accuracy_test  = model.score(X_test, y_test)
```

```python
print(f"Akurasi Model (Train) : {np.round(accuracy_train * 100,2)} %")
print(f"Akurasi Model (Test)  : {np.round(accuracy_test * 100,2)} %")
```
